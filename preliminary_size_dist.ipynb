```python
path_2ds_stats = '/gws/nopw/j04/dcmex/users/ezriab/processed_stats/2ds/all_2ds.csv'
path_labelled = '/gws/nopw/j04/dcmex/users/ezriab/image_labelling/2ds_10000_sample/2ds_10000_new_2nd.csv'

full_ds_stats = pd.read_csv(path_2ds_stats)
image_labelled_df = pd.read_csv(path_labelled)

images_labelled_lst = list(image_labelled_df['image_name'])

labelled_image_stats = full_ds_stats[full_ds_stats['name'].isin(images_labelled_lst)] # extract the ones labelled

image_labelled_df = image_labelled_df.rename(columns={'image_name': 'name'}) # need to rename our column, so they merge nicely
merged_df = pd.merge(labelled_image_stats, image_labelled_df, on='name', how='left')  # merge them together
```

```python
def reduce_full_df(big_df, label):
    smaller_df = big_df[
    (big_df['number_label'] == label) &
    (big_df['note'] != np.NaN)]
    #(~big_df.isnull().any(axis=1))]  # Exclude rows with NaN
    return smaller_df

def round_to_sf(x, sf):
    return np.round(x, sf - int(np.floor(np.log10(abs(x)))) - 1)
```

```python
# adding new column of rounded data
merged_df['rounded_Euclidean_d'] = merged_df['Euclidean_d_max'].apply(lambda x: round_to_sf(x, 3))

cp_df = reduce_full_df(merged_df,1)
cp_df_sorted = cp_df.sort_values(by=['rounded_Euclidean_d'])
```
